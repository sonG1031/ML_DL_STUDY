# 점진적 학습
- 훈련한 모델을 버리지 않고 새로운 데이터에 대해서만 조금씩 더 훈련하는 방식
- 대표적인 점진적 학습 알고리즘은 **확률적 경사 하강법**(Stochastic Gradient Descent)

### 경사 하강법의 훈련 과정
- 가장 가파른 길을 찾아 조금씩 내려옴

# 확률적 경사 하강법
- 확률적 : '무작위하게' 혹은 '랜덤하게'의 기술적 표현
- 경사 : 기울기를 말하는 것
- 하강법 : 내려가는 방법
    - 즉, 경사 하강법은 경사를 따라 내려가는 방법
- 확률적 경사 하강법 : 훈련 세트에서 샘플 하나씩 꺼내 손실 함수의 경사를 따라 최적의 모델을 찾는 알고리즘.

### 훈련 과정

1.   훈련 세트에서 랜덤하게 하나의 샘플을 선택하여 경사를 조금 내려감.
2.   그다음 훈련 세트에서 랜덤하게 또 다른 샘플을 하나 선택하여 경사를 조금 내려감.
3. 이런 식으로 전체 샘플을 다 사용할 때까지 계속함.
4. 만족할 위치에 도달하지 못하면 훈련 세트에 모든 샘플을 다시 채움.  
5. 만족할때까지 1, 2, 3, 4 반복

**에포크** : 경사 하강법에서 훈련 세트를 한 번 모두 사용하는 과정

### 미니배치 경사 하강법(minibatch gradient descent)
- 여러개의 샘플을 사용해 경사 하강법을 수행하는 방식
- 실전에서 많이 사용함.

### 배치 경사 하강법(batch gradient descent)
- 전체 샘플을 사용해 경사 하강법을 수행하는 방식
- 가장 안정적, 그러나 전체 샘플을 사용하면 그만큼 컴퓨터 자원을 많이 사용함.

### 확률적 경사 하강법과 신경망 알고리즘
확률적 경사 하강법을 꼭 사용하는 알고리즘이 바로 신경망 알고리즘이다.  
신경망은 일반적으로 많은 데이터를 사용하기 때문에 한 번에 모든 데이터를 사용하기 어렵다.  
또 모델이 매우 복잡하므로 수학적인 방법으로 해답을 얻기 힘들다.  
꼭 기억해야할 것 : 신경망 모델은 확률적 경사 하강법이나 미니배치 경사 하강법을 사용한다.

## 그렇다면 무엇의 경사를 내려가는가? 
그것은 바로 손실함수(loss function)

# 손실 함수(loss function)
- 확률적 경사 하강법이 최적화할 대상
- 어떤 문제에서 머신러닝 알고리즘이 얼마나 엉터리인지를 측정하는 기준
- 따라서 손실 함수의 값이 작을수록 좋음.

### 비용 함수(cost function)
- 손실 함수는 샘플 하나에 대한 손실을 정의하고
- 비용 함수는 훈련 세트에 있는 모든 샘플에 대한 손실 함수의 합을 말한다.
- 보통 둘을 엄격하게 구분하지 않고 섞어서 사용함.

### 분류 문제에서 손실 함수?
- 분류에서 손실은 정답을 못 맞히는 것.
- 그렇다면 정확도를 손실 함수로 사용할 수 있을까?
    - 손실 함수는 미분이 가능해야 하지만 정확도를 함수로 사용한다면 미분이 불가능함.
    - 따라서 연속적인 손실 함수가 필요함.

# 손실 함수는 미분이 가능해야하는 이유? (참고자료 참고)
손실 함수를 하강하는 과정에서 접선의 기울기를 통해 최적화를 진행하게 되는데  
함수의 접선의 기울기를 쉽게 구할려면 미분(도함수~)하여 구할 수 있음. 그래서 미분가능해야 하는 거임.

### 그럼 어떻게 연속적인 손실 함수를 만들 수 있을까?
- 로지스틱 회귀에서 로지스틱 회귀 모델이 확률을 출력함.
- 확률은 0\~1 사이의 어떤 값도 될 수 있기 때문에 연속적임.

# 로지스틱 손실 함수
- 샘플 4개의 예측 확률을 0.9, 0.3, 0.2, 0.8이라고 가정하고 손실 함수를 만드는 법을 알아보자

- 타깃이 양성 클래스 (타깃 = 1)
    - **-(예측 \* 타깃)**으로 계산하면 예측이 타깃에 가까울수록 값이 작아짐. 따라서 손실이 낮음.
        - ex) -(0.9 * 1) = -0.9
    - 반대로 예측이 타깃과 거리가 멀다면 값이 상대적으로 큼. 따라서 손실이 높음.
        - ex) -(0.3 * 1) = -0.3
- 타깃이 음성 클래스 (타깃 = 0)
    - 타깃이 0이므로 위처럼 계산하면 무조건 0이 나옴.
    - 따라서 타깃을 마치 양성 클래스처럼 바꾸어 1로 만듬.
    - 대신 예측 값도 양성 클래스에 대한 예측으로 바꿈.
    - 즉, **-((1 - 예측) \* 타깃)**으로 계산함.
        - ex) -(1 - 0.2) * 1) = -(0.8 * 1) = -0.8, 손실이 낮음.
        - ex) -(1 - 0.8) * 1) = -(0.2 * 1) = -0.2, 손실이 높음.

여기에서 예측 확률에 로그함수를 적용하면 더 좋다.  
- 타깃이 양성 클래스 (타깃 = 1)
    - -log(예측)
- 타깃이 음성 클래스 (타깃 = 0)
    - -log(1 - 예측)  

이유는?  (아래 그래프 참고)  
예측확률의 범위는 0~1 사이인데 로그 함수는 이 사이에서 음수가 되므로 최종 손실 값은 양수가 됨.  
따라서 손실이 양수이면 이해하기 더 쉽고, -log함수는 0에 가까워질수록 손실을 아주 크게 만들어 모델에 큰 영향을 미침.

이렇게 예측 확률을 전처리하여 손실을 극대화 시킨 함수를  
**로지스틱 손실 함수** 또는 **이진 크로스엔트로피 손실 함수**라고 부른다.

이름에서부터 티가 나지만 로지스틱 손실 함수를 사용하면 로지스틱 회귀 모델이 만들어짐.  
위의 예들은 이진 분류이지만 다중 분류도 매우 비슷한 손실 함수를 사용함.  
다중 분류에서 사용하는 손실 함수를 **크로스엔트로피 손실 함수**라고 부름.  
따라서 이진 분류에서는 로지스틱 손실 함수를 사용하고 다중 분류에서는 크로스엔트로피 손실 함수를 사용함.

### 그럼 회귀에는 어떤 손실 함수를 사용할까?
회귀의 손실 함수로 평균 절댓값 오차를 사용할 수 있음.  
타깃에서 예측을 뺀 절댓값을 모든 샘플에 평균한 값임.  
평균 제곱 오차도 많이 사용함. 타깃에서 예측을 뺀 값을 제곱한 다음 모든 샘플에 평균한 값임.  
이 값이 작을수록 좋은 모델이겠죠?ㅎㅎ

```python
# 데이터 준비
import pandas as pd
fish  = pd.read_csv("https://bit.ly/fish_csv_data")

fish_input = fish[['Weight', 'Length', 'Diagonal', 'Height', 'Width']].to_numpy()
fish_target = fish['Species'].to_numpy()
```

```python
# 전처리 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

train_input, test_input, train_target, test_target = train_test_split(fish_input, fish_target, random_state=42)

ss = StandardScaler()
ss.fit(train_input) # 통계 값 계산
train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input) # 훈련세트의 통계 값으로 테스트세트를 변환해야함!
```

## SGDClassifier
- 확률적 경사 하강법을 제공하는 대표적은 분류용 클래스
- 객체를 만들 때 2개의 매개변수를 지정함.
    - loss는 손실 함수의 종류를 지정
    - max_iter는 수행할 에포크 횟수를 지정



```python
from sklearn.linear_model import SGDClassifier 

# 로지스틱 손실 함수로 지정함.
sc = SGDClassifier(loss='log', max_iter=10, random_state=42)
sc.fit(train_scaled, train_target)
print(sc.score(train_scaled, train_target))
print(sc.score(test_scaled, test_target))
```

ConvergenceWarning  
모델이 충분히 수렴하지 않았을때 뜨는 경고

partial_fit() 
- 모델을 이어서 훈련할 때 사용함.
- 호출할 때마다 1에포크씩 이어서 훈련할 수 있음.

```python
sc.partial_fit(train_scaled, train_target)
print(sc.score(train_scaled, train_target))
print(sc.score(test_scaled, test_target))
# 정확도가 향상됨.
```

### 에포크와 과대/과소적합
- 확률적 경사 하강법을 사용한 모델은 에포크 횟수에 따라 과소적합이나 과대적합이 될 수 있음.
- 에포크 횟수가 적으면 모델이 훈련 세트를 덜 학습함.
- 에포크 횟수가 많으면 훈련 세트를 완전히 학습함.
- 즉, 적은 에포크 횟수는 과소적합된 모델일 가능성이 높고
- 많은 에포크 횟수는 과대적합된 모델일 가능성이 높음.

### 조기 종료(early stopping)
- 과대적합이 시작하기 전에 훈련을 멈추는 것

#### partial_fit 메서드의 classes 매개변수  
partial_fit 메서드는 일부 데이터를 사용해서 훈련할 때 사용하는 메서드입니다.  
훈련할 때마다 다른 타깃이 전달되면 일관된 모델을 만들 수 없습니다.
따라서 전체 클래스의 레이블을 전달해 주어야함.

```python
import numpy as np
sc = SGDClassifier(loss='log', random_state=42)
train_score = []
test_score = []
classes = np.unique(train_target)
```

```python
for _ in range(0, 300): # 300번의 에포크 
    sc.partial_fit(train_scaled, train_target, classes=classes)
    train_score.append(sc.score(train_scaled, train_target))
    test_score.append(sc.score(test_scaled, test_target))
```

```python
import matplotlib.pyplot as plt

plt.plot(train_score)
plt.plot(test_score)
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.show()
```

100번째 에포크가 적절한 반복 횟수로 보임.

```python
sc = SGDClassifier(loss='log', max_iter=100, tol=None, random_state=42)
sc.fit(train_scaled, train_target)
print(sc.score(train_scaled, train_target))
print(sc.score(test_scaled, test_target))
```

SGDClassifier는 일정 에포크 동안 성능이 향상되지 않으면 더 훈련하지 않고 자동으로 멈춤.  
tol 매개변수에서 향상될 최솟값을 지정함. tol=None으로 하여 자동으로 멈추지 않고 max_iter=100만큼 무조건 반복하도록 만듬.

### 참고 | SGDClassifier의 loss 매개변수  
loss 매개변수의 기본값은 'hinge'임. hinge loss는 서포트 벡터 머신(svm)이라 불리는 또다른 머신러닝 알고리즘을 위한 손실 함수
